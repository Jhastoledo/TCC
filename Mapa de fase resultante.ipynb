```python
#Para obter o mapa de fase resultante basta subtrair os mapas na condição de deformação e não deformação 
```

```python
# Importando as bibliotecas:

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img1 = cv.imread("mapa de fase 1.png", 0) #Mapa de fase sem deformação
img2 = cv.imread("mapa de fase 2.png", 0) #Mapa de fase com deformação
```

```python
sub1 = cv.subtract(img2,img1)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub1, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```
