```python
#A equação (4.2) diz que pode-se obter uma imagem utilizando a seguinte operação:
#Imagem = I3 - I1 / I0 - I2

#Vamos obter a imagem em 3 etapas:
#1ª etapa: calcular I3 - I1
#2ª etapa: calcular I0 - I2
#3ª etapa: fazer a divisão
```

```python
# Importando as bibliotecas:

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Importando as imagens usando a biblioteca do OpenCV:

I0 = cv.imread("imagem 0,7.png", 0) #Imagem de referência
I1 = cv.imread("imagem 0,8.png", 0) #Imagem rotacionada 0,1 grau no sentido anti-horário
I2 = cv.imread("imagem 0,9.png", 0) #Imagem rotacionada 0,2 grau no sentido anti-horário
I3 = cv.imread("imagem 1.png", 0) #Imagem rotacionada 0,3 grau no sentido anti-horário
```

```python
#1ª etapa: I3 - I1

num = cv.subtract(I3,I1)
fig = plt.figure(figsize=(40, 20))
plt.imshow(num, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
#2ª etapa: I0 - I2

den = cv.subtract(I3,I1)
fig = plt.figure(figsize=(40, 20))
plt.imshow(den, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
#3ª etapa: num/den

im = cv.divide(num, den)
fig = plt.figure(figsize=(40, 20))
plt.imshow(den, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```
