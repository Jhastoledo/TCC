```python
# Neste documento será realizada a subtração dos padrões de Speckle. As imagens subtraidas vão da "Imagem 0" até a "Imagem 509D", 
# com a variação de 1px. Neste documento será detalhado todo o processo.
```

```python
# Imagem 511D: Imagem transladada 1px pra baixo em relação a imagem 0 - =Imagem 516U
# Imagem 510D: Imagem transladada 2px pra baixo em relação a imagem 0 - =Imagem 515U
```

```python
# Importando as bibliotecas:

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Importando as imagens usando a biblioteca do OpenCV:

img0 = cv.imread("imagem 0.png", 0) #Imagem de referência
img1 = cv.imread("imagem 511D.png", 0) #Imagem transladada 1px p baixo
img2 = cv.imread("imagem 510D.png", 0) #Imagem transladada 2px p baixo
```

```python
# Fazendo a operação de subtração nas imagens. Sempre iremos tomar uma imagem de referência e outra como sendo deformada.
# A operação de subtração é feita pela biblioteca do OpenCV
# A imagem é plotada depois da operação de subtração usando a biblioteca matplotlib
```

```python
# Subtração da "Imagem 0" (referência) - "Imagem 511D" (deformada).

sub1 = cv.subtract(img0,img1)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub1, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
# Subtração da "Imagem 511D" (referência) - "Imagem 510D" (deformada).

sub2 = cv.subtract(img1,img2)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub2, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```
