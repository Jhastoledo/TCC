```python
# Neste documento será realizada a subtração dos padrões de Speckle. As imagens subtraidas vão da "Imagem 0" até a "Imagem 517U", 
# com a variação de 1px. Neste documento será detalhado todo o processo.
```

```python
# Imagem 513U: Imagem transladada 1px pra cima em relação a imagem 0
# Imagem 514U: Imagem transladada 2px pra cima em relação a imagem 0
# Imagem 515U: Imagem transladada 3px pra cima em relação a imagem 0
# Imagem 516U: Imagem transladada 4px pra cima em relação a imagem 0
# Imagem 517U: Imagem transladada 5px pra cima em relação a imagem 0  = Imagem0 - a partir daqui (5px) fica ciclico
```

```python
# Importando as bibliotecas:

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Importando as imagens usando a biblioteca do OpenCV:

img0 = cv.imread("imagem 0.png", 0) #Imagem de referência
img1 = cv.imread("imagem 513U.png", 0) #Imagem transladada 1px p cima
img2 = cv.imread("imagem 514U.png", 0) #Imagem transladada 2px p cima
img3 = cv.imread("imagem 515U.png", 0) #Imagem transladada 3px p cima
img4 = cv.imread("imagem 516U.png", 0) #Imagem transladada 4px p cima
img5 = cv.imread("imagem 517U.png", 0) #Imagem transladada 5px p cima
```

```python
# Fazendo a operação de subtração nas imagens. Sempre iremos tomar uma imagem de referência e outra como sendo deformada.
# A operação de subtração é feita pela biblioteca do OpenCV
# A imagem é plotada depois da operação de subtração usando a biblioteca matplotlib
```

```python
# Subtração da "Imagem 0" (referência) - "Imagem 513U" (deformada).

sub1 = cv.subtract(img0,img1)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub1, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
# Subtração da "Imagem 513U" (referência) - "Imagem 514U" (deformada).

sub2 = cv.subtract(img1,img2)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub2, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
# Subtração da "Imagem 514U" (referência) - "Imagem 515U" (deformada).

sub3 = cv.subtract(img2,img3)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub3, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
# Subtração da "Imagem 515U" (referência) - "Imagem 516U" (deformada).

sub4 = cv.subtract(img3,img4)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub4, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```

```python
# Subtração da "Imagem 516U" (referência) - "Imagem 517U" (deformada).

sub5 = cv.subtract(img4,img5)
fig = plt.figure(figsize=(40, 20))
plt.imshow(sub5, cmap= plt.cm.gray)
plt.axis('off') #parar de mostrar os eixos
plt.show()
```
